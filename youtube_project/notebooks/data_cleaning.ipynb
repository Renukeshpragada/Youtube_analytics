```python
import pandas as pd

channel_df = pd.read_csv("../data/raw/channel_raw.csv")
videos_df = pd.read_csv("../data/raw/videos_raw.csv")
```

```python
videos_df.info()
videos_df.isnull().sum()
```

```python
videos_df["published_date"] = pd.to_datetime(videos_df["published_date"])

videos_df["views"] = videos_df["views"].astype(int)
videos_df["likes"] = videos_df["likes"].astype(int)
videos_df["comments"] = videos_df["comments"].astype(int)
```

```python
videos_df = videos_df.drop_duplicates(subset="video_id")
```

```python
videos_df["engagement_rate"] = (
    (videos_df["likes"] + videos_df["comments"]) / videos_df["views"]
).round(4)
```

```python
top_videos_df = videos_df.sort_values(
    by="views", ascending=False
).head(10)
```

```python
videos_df["upload_month"] = videos_df["published_date"].dt.to_period("M")

upload_frequency = videos_df.groupby("upload_month").size().reset_index(name="video_count")
upload_frequency
```

```python
videos_processed = videos_df[
    [
        "video_id",
        "channel_id",
        "title",
        "published_date",
        "views",
        "likes",
        "comments",
        "engagement_rate"
    ]
]
```

```python
channel_df.to_csv("../data/processed/channel_processed.csv", index=False)
videos_processed.to_csv("../data/processed/videos_processed.csv", index=False)
top_videos_df.to_csv("../data/processed/top_videos.csv", index=False)

print("Week-2 cleaned & transformed data saved successfully")
```
